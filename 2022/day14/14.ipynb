```python
# Part 1

with open('input.txt') as f:
    lines = f.read().strip().split('\n')

# Find the bounds and draw the field
min_x = 500
max_x = 0
max_y = 0

for line in lines:
    coords = line.split(" -> ")
    for coord in coords:
        x, y = [int(i) for i in coord.split(',')]
        if x < min_x:
            min_x = x
        if x > max_x:
            max_x = x
        if y > max_y:
            max_y = y

field = [['.' for q in range(max_x + 1 - min_x)] for p in range(max_y + 1)]

# Draw the rock structure in the field
for line in lines:
    coords = line.split(" -> ")
    for i in range(len(coords) - 1):
        x0, y0 = [int(j) for j in coords[i].split(',')]
        x1, y1 = [int(j) for j in coords[i + 1].split(',')]

        for x in range(min(x0, x1) - min_x, max(x0, x1) - min_x + 1):
            for y in range(min(y0, y1), max(y0, y1) + 1):
                field[y][x] = '#'

# Drop the sand
def drop_sand():
    grains = 0
    while True:
        x = 500 - min_x
        y = 0
        while True:
            if (y + 1) > max_y:
                return grains
            if field[y + 1][x] == '.':
                y += 1
                continue
            else:
                if (x - 1) < 0:
                    return grains
                if field[y + 1][x - 1] == '.':
                    x -= 1
                    y += 1
                    continue

                if (x + 1 + min_x) > max_x:
                    return grains
                if field[y + 1][x + 1] == '.':
                    x += 1
                    y += 1
                    continue

                field[y][x] = 'o'
                grains += 1
                break

grains = drop_sand()

for line in field:
    print(''.join(line))

print(grains)
```

```python
# Part 2

with open('input.txt') as f:
    lines = f.read().strip().split('\n')

# Find the bounds and draw the field
min_x = 500
max_x = 0
max_y = 0

for line in lines:
    coords = line.split(" -> ")
    for coord in coords:
        x, y = [int(i) for i in coord.split(',')]
        if x < min_x:
            min_x = x
        if x > max_x:
            max_x = x
        if y > max_y:
            max_y = y

max_x += 200

field = [['.' for q in range(max_x + 1)] for p in range(max_y + 1)]

# Draw the rock structure in the field
for line in lines:
    coords = line.split(" -> ")
    for i in range(len(coords) - 1):
        x0, y0 = [int(j) for j in coords[i].split(',')]
        x1, y1 = [int(j) for j in coords[i + 1].split(',')]

        for x in range(min(x0, x1), max(x0, x1) + 1):
            for y in range(min(y0, y1), max(y0, y1) + 1):
                field[y][x] = '#'

field.append(['.' for r in range(max_x + 1)])
field.append(['#' for r in range(max_x + 1)])

max_y += 2

# Drop the sand
def drop_sand():
    grains = 0
    while True:
        x = 500
        y = 0
        # Return if the grains reach the top
        if field[y][x] == 'o':
            return grains

        while True:

            if field[y + 1][x] == '.':
                y += 1
                continue
            else:

                if field[y + 1][x - 1] == '.':
                    x -= 1
                    y += 1
                    continue

                if field[y + 1][x + 1] == '.':
                    x += 1
                    y += 1
                    continue

                field[y][x] = 'o'
                grains += 1
                break

grains = drop_sand()

for line in field:
    print(''.join(line[350:600]))

print(grains)
```
